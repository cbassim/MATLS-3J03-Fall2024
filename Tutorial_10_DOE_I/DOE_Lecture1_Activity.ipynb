```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Yogurt

```python
X1=pd.Series([-1,1,-1,1],name='Fat')
X2=pd.Series([-1,-1,1,1], name='Time')
X3=pd.Series([1,-1,-1,1],name="Interaction")
Y=pd.Series([5,8,6,9],name='Output')
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

DF=pd.concat([X1,X2,X3,Y],axis='columns')
print(DF)
```

```python
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Fat",y="Output",data=DF,ax=ax[0],errorbar=None)
sns.lineplot(x="Time",y="Output",data=DF, ax=ax[1],errorbar=None)
```

```python
# factorplot, catplot
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Fat",y="Output",hue="Time",data=DF,ax=ax[0])
sns.lineplot(x="Time",y="Output",hue="Fat",data=DF, ax=ax[1])
```

```python
model1=ols('Y~X1+X2',data=DF)
results1=model1.fit()
print(results1.summary())
```

# Interaction effects

```python
Xa=pd.Series([-1,1,-1,1],name='Temp')
Xb=pd.Series([-1,-1,1,1], name='Substrate')
Xc=pd.Series([1,-1,-1,1])
Yabc=pd.Series([39,80,61,82],name='Output')
```

```python
DFabc=pd.concat([Xa,Xb,Xc,Yabc],axis='columns')
print(DFabc)
```

```python
model3=ols('Yabc~Xa+Xb+Xa*Xb',data=DFabc)
results3=model3.fit()
print(results3.summary())
```

```python
# factorplot, catplot
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Temp",y="Output",hue="Substrate",data=DFabc,ax=ax[0])
sns.lineplot(x="Substrate",y="Output",hue="Temp",data=DFabc, ax=ax[1])
```
