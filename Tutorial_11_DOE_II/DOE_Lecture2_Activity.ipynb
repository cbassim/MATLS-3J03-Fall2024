```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
X1=pd.Series([-1,1,-1,1],name='Soil')
X2=pd.Series([-1,-1,1,1], name='Water')
X3=pd.Series([1,-1,-1,1],name="Interaction")
Y=pd.Series([39,50,61,82],name='Output')
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm

DF=pd.concat([X1,X2,X3,Y],axis='columns')
print(DF)
```

```python
# factorplot, catplot
fig, ax =plt.subplots(1,2,sharex=True)
sns.lineplot(x="Soil",y="Output",hue="Water",data=DF,ax=ax[0])
sns.lineplot(x="Water",y="Output",hue="Soil",data=DF, ax=ax[1])
```

```python
model1=ols('Y~X1+X2',data=DF)
results1=model1.fit()
print(results1.summary())
```

```python
X1=np.array([-1,1,-1,1])
X2=np.array([-1,-1,1,1])
X3=np.array([1,-1,-1,1])
Y=np.array([39,50,61,82])

model2=ols('Y~X1+X2+X1*X2',data=DF)
results2=model2.fit()
print(results2.summary())
```

```python
DF
```

```python
fig=plt.figure(figsize=plt.figaspect(1)*2)
ax=fig.add_subplot(111,projection='3d')

df = DF
x = df["Soil"]
y = df["Water"]
z = df["Output"]

x1,x2,x3=model2.exog_names[1:]

#x_surf=np.arange(x.min(), x.max(),1)
x_surf=np.arange(-1,1,.1)
y_surf=np.arange(-1,1,.1)
#y_surf=np.arange(y.min(), y.max(),.1)
xS,yS=np.meshgrid(x_surf,y_surf)

exog=pd.DataFrame({x1: xS.ravel(), x2:yS.ravel()})
zS=results2.predict(exog=exog).values.reshape(xS.shape)
ax.plot_surface(xS,yS,zS,alpha=0.4,cmap='coolwarm')

ax.scatter(x,y,z,marker='*',color="g",label="Raw Data")
ax.view_init(elev=1,azim=230)

ax.scatter(x.values,y.values,results2.fittedvalues.values,label="Fited Values",color="r")

plt.xlabel("Soil")
plt.ylabel("Water")
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

CountourA=ax.contourf(xS,yS,zS,cmap="coolwarm")
fig.colorbar(CountourA,shrink=0.5, aspect=5)
plt.xlabel("Soil")
plt.ylabel("Water")
```
