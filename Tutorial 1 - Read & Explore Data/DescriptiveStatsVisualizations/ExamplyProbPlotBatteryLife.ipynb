```python
%reset
# Importing Necessary Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
BatteryLife = pd.Series([176,183,185,190,191,192,201,205,214,220])
BatteryLife.index=np.arange(1, len(BatteryLife)+1)
BatteryLife
```

```python
# In Montgomery book
j=np.arange(1,11)
perc=(j-.5)/10
perc
```

```python
# Or through percentiles
k=np.arange(0,1.1,.1)
```

```python
np.percentile(k,([5,15,25,35,45,55,65,75,85,95]))
```

```python
sns.scatterplot(x=BatteryLife, y=(perc*100))
plt.xlabel('x(j)= data value')
plt.ylabel('100(j-0.5)/n')
sns.lineplot(x=[176,220], y=[0,100])
```

```python
# the z-score or standard normal ppf (point percent function) of the percentiles of normal distribution
from scipy.stats import norm
z=norm.ppf(perc)
z
```

```python
sns.scatterplot(x=BatteryLife, y=z)
plt.xlabel('x(j)= data value')
plt.ylabel('zj')
```
