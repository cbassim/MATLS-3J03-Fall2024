```python
%reset
# Importing Necessary Libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# load in the "penguins" dataset that can be loaded from seaborn;
# another great scatterplot dataset is R's iris dataset
penguins=sns.load_dataset("penguins")
```

```python
penguins
```

```python
# matplotlib used plot to make a scatterplot
plt.plot("bill_length_mm","flipper_length_mm",data=penguins,linestyle='none',marker='o')
plt.xlabel('Bill Length mm')
plt.ylabel("Flipper Length mm")
```

```python
# let's go ahead and work with seaborn
# lets set the theme we want seaborn to use
sns.set_theme(style='ticks')
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",data=penguins)
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",hue="species",style="species",data=penguins)
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",hue="body_mass_g",style="body_mass_g",sizes=(20,200), legend=False,data=penguins)
```

```python
#relplot() allows groupings
sns.relplot(x="bill_length_mm",y="flipper_length_mm",col="species",data=penguins)
```

```python
#regplot() plots data and a linear regression line
sns.regplot(x="bill_length_mm",y="flipper_length_mm",data=penguins)
```

```python
# lmplot() combines regplot and facetgrid to fit regression lines across conditional subsets of data
sns.lmplot(x="bill_length_mm",y="flipper_length_mm",hue="species",data=penguins)
```

```python
# pairplot() plot pairwise relationships:  Scatterplot Matrix
sns.pairplot(penguins, hue='species')
```

```python
# jointplot() to draw a marginal plot:  marginal plots show the relationship between 2 nueric variables, the cetral chart shows the correlation scatterplot
sns.jointplot(x="bill_length_mm",y="flipper_length_mm",kind="scatter",data=penguins)
```

```python
# central plot custum style:  'scatter', 'reg', 'resid','kde', 'hex'
sns.jointplot(x="bill_length_mm",y="flipper_length_mm",kind="hex",data=penguins)
```

```python
# custimize the jointplot marginal chats, showing the distribution of the two variables
sns.jointplot(x="bill_length_mm",y="flipper_length_mm",kind="scatter",marginal_kws=dict(bins=30),data=penguins)
```

```python
# adding hue will add color legend adn draw seperate density curves to the margins (using kdeplot())
sns.jointplot(x="bill_length_mm",y="flipper_length_mm",hue='species',data=penguins)
```
