# Empirical Rule

This program demonstrates the empirical rule, by:

* Generating a set of empirical data based on the provided mean and standard deviation.
* Plotting this data as a histogram to show its distribution.
* Overlaying the normal distribution curve to illustrate the theoretical distribution.
* Highlighting areas within one, two, and three standard deviations from the mean to demonstrate the Empirical Rule.

When you run this code, you'll be prompted to enter a mean and standard deviation. After entering these values, the function generates and displays a histogram of the empirical data, overlays the normal distribution curve, and marks the ranges for the Empirical Rule, providing a clear and comprehensive visualization of how the empirical data compares to the theoretical model.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to plot normal distribution, empirical rule, and histogram of generated data
def plot_distribution_with_histogram(mu, sigma):
    # Generate normally distributed data
    data = np.random.normal(mu, sigma, 1000)
    
    # Generate a range of values for the normal distribution curve
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    # Probability density function for the normal distribution
    p = norm.pdf(x, mu, sigma)
    
    # Plotting the histogram of the empirical data
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='gray', label='Empirical Data')
    
    # Plotting the normal distribution curve
    plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
    
    # Fill under the curve for each standard deviation range
    plt.fill_between(x, p, where=((x >= mu - sigma) & (x <= mu + sigma)), color='skyblue', alpha=0.5, label='Within 1 STD (68%)')
    plt.fill_between(x, p, where=((x >= mu - 2*sigma) & (x < mu - sigma)) | ((x > mu + sigma) & (x <= mu + 2*sigma)), color='mediumseagreen', alpha=0.5, label='Within 2 STD (95%)')
    plt.fill_between(x, p, where=((x >= mu - 3*sigma) & (x < mu - 2*sigma)) | ((x > mu + 2*sigma) & (x <= mu + 3*sigma)), color='salmon', alpha=0.5, label='Within 3 STD (99.7%)')
    
    # Adding titles and labels
    plt.title('Normal Distribution with Empirical Data - Empirical Rule')
    plt.xlabel(f'Mean = {mu}, Standard Deviation = {sigma}')
    plt.ylabel('Probability Density')
    plt.legend()
    
    plt.show()

# Example usage
mu = float(input("Enter the mean (e.g., 0): "))
sigma = float(input("Enter the standard deviation (e.g., 1): "))
plot_distribution_with_histogram(mu, sigma)
```
